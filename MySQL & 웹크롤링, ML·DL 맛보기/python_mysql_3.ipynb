```python
sql = 'create database if not exists best_seller;'
cursor.execute(sql)
```

```python
sql = 'use best_seller;'
cursor.execute(sql)
```

```python
sql = '''create table best_seller (
ranking int,
title varchar(100),
url varchar(200),
price varchar(20)
);'''

cursor.execute(sql)
```

```python
for li in lis:
    ranking = li.select('p')[0].text
    title  = li.select('a.itemname')[0].text
    url = li.select('a.itemname')[0]['href']
    price = li.select('div.item_price > div.s-price > strong')[0].text
    
    print(ranking)
    print(title)
    print(url)
    print(price)
    print()
```

```python
for li in lis:
    ranking = li.select('p')[0].text
    title  = li.select('a.itemname')[0].text
    url = li.select('a.itemname')[0]['href']
    price = li.select('div.item_price > div.s-price > strong')[0].text
    
    sql = "insert into best_seller values ({}, '{}', '{}', '{}');".format(ranking, title, url, price)
    
    cursor.execute(sql)
```

```python
sql = 'select * from best_seller;'
df_best_seller = pd.read_sql(sql, db)
```

```python
df_best_seller.to_csv('gmarket_best_seller.csv')
```
