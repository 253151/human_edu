## 네이버 싸이트 크롤링 간단 예제
- 상태코드 확인 : https://www.whatap.io/ko/blog/40/
- 200 : 정상적으로 작동
- 404 : 주소 입력이 잘못됨 (주소가 사라지는 경우)
- 503 : 회사 서버가 내려간 상태

```python
import requests
from bs4 import BeautifulSoup

url  = 'https://www.naver.com/'

# url 변수에 담긴 url(네이버) html 문서를 출력 
req = requests.get(url)
print(req.status_code)
print(req.text)
```

### 네이버 헤드 뉴스 찾기
- 1단계 : 사이트 정상인지 확인
  + `.py` 만들고, 배치로 돌려야 함. 

```python
import requests
from bs4 import BeautifulSoup 

# 데이터 수집
def crawling(soup):
  # print(soup)
  result = []
  
  # div 태그명 list_issue
  div = soup.find('div', class_ = 'list_issue')
  print(div)
  print('-----')

  for a in div.find_all('a'):
    print(a.get_text())
    result.append(a.get_text())

  return result

# 데이터 가공 (향후 프로젝트 진행 시)

# 데이터 DB에 저장 (향후 프로젝트 진행 시)

def main():
  # 코드 작성 후, crawling 함수에 객체 전달 (soup)
  custom_header = {
      'referer' : 'https://www.naver.com/', 
      'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.103 Safari/537.36'
  }

  req = requests.get(url, headers = custom_header)

  # soup 객체 
  soup = BeautifulSoup(req.text, "html.parser")
  result = crawling(soup)
  print('---최종결과---')
  print(result)

if __name__ == "__main__":
  main()
```
