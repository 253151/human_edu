## 새로운 컬럼 추가
- 오렌지와 레몬 판매량 합계

```python
data['판매량']= data['Lemon'] + data['Orange']
data.head()
```

```python
# 매출액
data['매출액'] = data['Price'] * data['판매량']
data.head()
```
